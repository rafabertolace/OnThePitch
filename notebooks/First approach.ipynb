```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 2, 1)
plt.scatter(df_rizespor['AwayTeam'], df_rizespor['B365>2.5'], label = 'Opening');
plt.scatter(df_rizespor['AwayTeam'], df_rizespor['B365C>2.5'], label = 'Closing');
plt.legend();
plt.subplot(1, 2, 2)
plt.scatter(df_rizespor['Date'], df_rizespor['B365>2.5'], label = 'Opening');
plt.scatter(df_rizespor['Date'], df_rizespor['B365C>2.5'], label = 'Closing');
plt.legend();
```
